 Create a new column - 'Bins_Hp'
where horsepower is
(70, 100]
(100, 200]
(200, 300]
(300, 400]
 >400

```python
def Bin_Hp(x):
    if x> 400:
        return '400'
    elif x > 300:
        return '(300, 400]'
    elif x > 200:
        return '(200, 300]'
    elif x> 100:
        return '(100, 200]'
    else :
        return '(70, 100]'
    
```

```python
Bin_Hp(300)
```

```python
Bin_Hp(200)
```

```python
df_cars['Bin_Hp']=df_cars.Horsepower.apply(Bin_Hp)
```
