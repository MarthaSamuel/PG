## Data  Standardization

```python
def standardize_test(test):
    return (test - test.mean()) / test.std()
```

```python
standardize_test(df_scores['test1'])
```

```python
standardize_test(df_scores)
```

```python
def standardize_test_table(df):
    return df_scores.apply(standardize_test)
```
